```python
Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
```

ANS-Seaborn is a popular Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics. Here are five common plots that you can create using Seaborn and their uses:

Scatter Plot:
Use: Scatter plots display individual data points as dots, allowing you to visualize the relationship between two numerical variables. They are useful for identifying patterns, clusters, or trends in data.

Line Plot:
Use: Line plots are used to visualize the trend of a numerical variable over time or any other continuous interval. They are commonly used for time-series data or data with a natural ordering.

Bar Plot:
Use: Bar plots are suitable for comparing categorical data. They display the distribution of a categorical variable and allow you to compare values across different categories easily.

Histogram:
Use: Histograms are used to represent the distribution of a single numerical variable. They show how data is spread across various bins, helping to identify the underlying data distribution and potential outliers.

Box Plot (Box-and-Whisker Plot):
Use: Box plots are useful for visualizing the distribution of numerical data through their quartiles. They display the median, interquartile range, and potential outliers, making them effective for comparing multiple data groups.

Seaborn offers a user-friendly and aesthetically pleasing approach to create these plots with minimal code, making it a valuable tool for data exploration and presentation.

```python
Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.
```

```python
ANS-

import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load the "fmri" dataset
fmri_data = sns.load_dataset("fmri")

# Step 3: Create the line plot
plt.figure(figsize=(10, 6))  # Optional: Set the figure size

# Use Seaborn's lineplot function to plot the line plot
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)

# Optional: Add title and labels to the plot
plt.title("FMRI Signal over Time for Different Events and Regions")
plt.xlabel("Timepoint")
plt.ylabel("Signal")

# Show the plot
plt.show()
```

```python
Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.
```

```python
ANS-

import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load the "titanic" dataset
titanic_data = sns.load_dataset("titanic")

# Step 3: Create the box plots
plt.figure(figsize=(10, 6))  # Optional: Set the figure size

# First box plot for 'age' and 'pclass'
sns.boxplot(x="pclass", y="age", data=titanic_data)
plt.title("Age Distribution across Passenger Classes")
plt.xlabel("Passenger Class")
plt.ylabel("Age")

plt.figure(figsize=(10, 6))  # Optional: Set the figure size

# Second box plot for 'fare' and 'pclass'
sns.boxplot(x="pclass", y="fare", data=titanic_data)
plt.title("Fare Distribution across Passenger Classes")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")

# Show the plots
plt.show()
```

```python
Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load the "diamonds" dataset
diamonds_data = sns.load_dataset("diamonds")

# Step 3: Create the histogram with 'cut' as hue
plt.figure(figsize=(10, 6))  # Optional: Set the figure size

# Use Seaborn's histplot function to create the histogram
sns.histplot(data=diamonds_data, x='price', hue='cut', kde=True)

# Optional: Add title and labels to the plot
plt.title("Price Distribution for Different Diamond Cuts")
plt.xlabel("Price")
plt.ylabel("Count")

# Show the plot
plt.show()
```

```python
Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load the "iris" dataset
iris_data = sns.load_dataset("iris")

# Step 3: Create the pair plot with 'species' as hue
plt.figure(figsize=(10, 8))  # Optional: Set the figure size

# Use Seaborn's pairplot function to create the pair plot
sns.pairplot(data=iris_data, hue='species')

# Show the plot
plt.show()
```
